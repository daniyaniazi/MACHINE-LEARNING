```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<font color="purple"><h2 align="center">  Importing the dataset</h2></font>


```python
dataset = pd.read_csv('C:\\Users\\DANIA NIAZI\\Desktop\\COURSES\\MACHINE_LEARNING\\REGRESSION\\Datsets\\dataset.csv')
dataset
```


<font color="purple"><h2 align="center"> Splitting the dataset into the Training set and Test set set</h2></font>

```python
x = dataset.iloc[:, 2:3].values
y = dataset.iloc[:, 3:4].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 1)
```



<font color="purple"><h2 align="center"> Fitting Simple Linear Regression to the Training set</h2></font>

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

 
<font color="purple"><h3 align="center"> Predicting the Test set results</h3></font>

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

<font color="purple"><h3 align="center"> Visualising the Training set results</h3></font>

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Head Size(cm^3) vs Brain Weight(grams)(Training set)')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.show()
```


<font color="purple"><h3 align="center"> Visualising the testing  set results</h3></font>

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Head Size(cm^3) vs Brain Weight(grams) (Test set)')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.show()
```
