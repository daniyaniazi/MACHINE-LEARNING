```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error,r2_score
```

```python
dataset = pd.read_csv('50_Startups.csv')
dataset.head()
```

# Seperating DataSet Acc To State

```python
dataset.set_index('State',inplace=True)
dataset
```

```python
New_York_df=pd.DataFrame(dataset.loc['New York'])
California_df=pd.DataFrame(dataset.loc['California'])
Florida_df=pd.DataFrame(dataset.loc['Florida'])
```

# APPLYING LINEAR REGRESSION TO DATASETS

## NEWYORK STATE

```python
New_York_df.reset_index(inplace=True)
New_York_df
```

```python
NY_x = New_York_df.iloc[:,1:-1].values
NY_y = New_York_df.iloc[:,-1].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
NY_x_train, NY_x_test, NY_y_train, NY_y_test = train_test_split(NY_x ,NY_y, test_size = 0.3, random_state = 0)
```

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
NY_lin_reg = LinearRegression()
NY_lin_reg.fit(NY_x_train, NY_y_train)
```

```python
# Predicting a new result
NY_lin_reg.predict([[1000.23,124153.04,1903.93]])
```

```python
NY_y_pred_lin =NY_lin_reg.predict( NY_x_test)
NY_y_pred_lin
```

```python
#R2_SCORE OF NY USING LINEAR REG
NY_lin_reg_acc=r2_score(NY_y_test,NY_y_pred_lin)
print(r2_score( NY_y_test,NY_y_pred_lin))
```
