##### Apriori Components
![image](params.PNG)
- <b> Support </b> : 
    - Default popularity of any item
    - support(movie1) = Transaction involving movei 1 / Total transaction
- <b> Confidence </b> :
    - Liklihood that the customer bought both movie 1 and movie 2 tickets
    - Confidence =Transaction involving both movie 1 and 2 /Transaction involving movie 1
- <b> lift </b> : 
    - lift : (confidence(m1->m2)/support(m1)
    - Likelihood of a customer watch both both movie 1 and 2 together is {lift}  times more than the chancde of purchasing movie 1 alone
    - lift < 1 : unlikely to buy both
    - greater the value , the better combination

```python
# Data Preprocessing
dataset = pd.read_csv('Market.csv', header = None)
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
# Training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions, 
                min_support = 0.003, #Product that sells out 3 to 4 times in a day => 3 * 7/ 7501 = 0.003 for a week
                min_confidence = 0.2,
                min_lift = 3,
                min_length = 2) # Binary pairing eg : bread + jam

```
